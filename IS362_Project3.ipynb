In this project, I am reading data from the Chinook SQL database into Python using pandas.  
I will connect to the database, write a SQL query that joins multiple tables, and display the results in a pandas DataFrame.

```python
import pandas as pd
import sqlite3

# I am connecting to the Chinook SQLite database file
conn = sqlite3.connect("data/Chinook_Sqlite.sqlite")

# Quick test to make sure the connection works
test_query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql_query(test_query, conn)

# Show the tables in the database
tables
```

## SQL query and DataFrame

For this part, I wrote a SQL query that joins the Customer, Invoice, InvoiceLine, Track, and Album tables.  
This lets me show each customer's name along with the tracks they purchased and the album title.

```python
# I am joining 5 tables to get customer names, track names, and album titles
query = """
SELECT
    c.LastName,
    c.FirstName,
    t.Name AS TrackName,
    a.Title AS AlbumTitle
FROM Customer c
JOIN Invoice i
    ON c.CustomerId = i.CustomerId
JOIN InvoiceLine il
    ON i.InvoiceId = il.InvoiceId
JOIN Track t
    ON il.TrackId = t.TrackId
JOIN Album a
    ON t.AlbumId = a.AlbumId
ORDER BY c.LastName, c.FirstName
"""

# Read the SQL query result into a pandas DataFrame
customer_tracks_df = pd.read_sql_query(query, conn)

# Show the first 5 rows
customer_tracks_df.head()
```

## Results

The query worked correctly and returned a pandas DataFrame with customer last names, first names, purchased track names, and album titles.  
I joined five tables in order to connect the customer information to the invoices, invoice lines, tracks, and albums.  
The results are sorted by customer last name and first name, as required.
